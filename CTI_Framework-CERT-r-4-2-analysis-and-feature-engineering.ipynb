```python
df_device = device_log.copy()
```

```python
import numpy as np
from datetime import datetime

connect_df = df_device[df_device["activity"] == "Connect"]
# Convert 'date' column to datetime
df_device['date'] = pd.to_datetime(df_device['date'], format='%m/%d/%Y %H:%M:%S')

# Extract date and hour
df_device['only_date'] = df_device['date'].dt.date
df_device['hour'] = df_device['date'].dt.hour

# 1. daily_device_connects: count per user per day (only "Connect" actions)
daily_device_connects = (
    df_device[df_device['activity'] == 'Connect']
    .groupby(['user', 'only_date'])
    .size()
    .reset_index(name='daily_device_connects')
)

# 2. avg_daily_device_connections: average connects per user per day
avg_device_use = (
    daily_device_connects
    .groupby('user')['daily_device_connects']
    .mean()
    .reset_index(name='avg_daily_device_connections')
)

# 3. deviation_from_avg_device_use: spike detection
# Join avg to daily
daily_with_avg = pd.merge(daily_device_connects, avg_device_use, on='user')
# Add deviation flag (e.g., spike = more than 2× avg)
daily_with_avg['deviation_from_avg_device_use'] = (
    daily_with_avg['daily_device_connects'] > 2 * daily_with_avg['avg_daily_device_connections']
)

# 4. late_hour_device_use: after-hours (e.g., >18:00 or <6:00)
df_device['after_hours'] = df_device['hour'].apply(lambda x: x < 6 or x > 18)
late_device_use = (
    df_device[(df_device['activity'] == 'Connect') & (df_device['after_hours'])]
    .groupby('user')
    .size()
    .reset_index(name='late_hour_device_use')
)

# 5. missing_disconnects: mismatched Connect/Disconnect counts per user
connect_counts = df_device[df_device['activity'] == 'Connect'].groupby('user').size()
disconnect_counts = df_device[df_device['activity'] == 'Disconnect'].groupby('user').size()
missing_disconnects = (
    (connect_counts - disconnect_counts).fillna(0).astype(int).reset_index(name='missing_disconnects')
)

# 6. first_use_date: first time user connected a device
first_use = (
    df_device[df_device['activity'] == 'Connect']
    .groupby('user')['date']
    .min()
    .reset_index(name='first_use_date')
)



distinct_pcs = connect_df.groupby("user")["pc"].nunique().reset_index(name="distinct_pcs_used")

# === Final Merge ===
# Merge all features on 'user'
features = avg_device_use.merge(first_use, on='user', how='left') \
                         .merge(late_device_use, on='user', how='left') \
                         .merge(missing_disconnects, on='user', how='left')\
                         .merge(distinct_pcs, on="user", how="left")  

# Fill NaNs where user had no after-hour usage or missing disconnects
features['late_hour_device_use'] = features['late_hour_device_use'].fillna(0).astype(int)
features['missing_disconnects'] = features['missing_disconnects'].fillna(0).astype(int)

# Optional: Save daily_with_avg if you want time-series per day
# daily_with_avg.to_csv('daily_device_usage.csv', index=False)

#======================== NEW FEATURE =======================
# One std + avg above device usage results in a flag!
device_use_std = (
    daily_device_connects
    .groupby('user')['daily_device_connects']
    .std()
    .reset_index(name='std_daily_device_connections')
)

# Join std to daily_with_avg
daily_with_avg = pd.merge(daily_with_avg, device_use_std, on='user', how='left')

# Compute threshold and flag
daily_with_avg['above_threshold'] = (
    daily_with_avg['daily_device_connects'] > 
    (daily_with_avg['avg_daily_device_connections'] + daily_with_avg['std_daily_device_connections'])
)

# Count number of days above the threshold per user
above_avg_daily_counter = (
    daily_with_avg[daily_with_avg['above_threshold']]
    .groupby('user')
    .size()
    .reset_index(name='above_avg_daily_counter')
)

# Merge into main features table
features = features.merge(above_avg_daily_counter, on='user', how='left')
features['above_avg_daily_counter'] = features['above_avg_daily_counter'].fillna(0).astype(int)

#============================= NEW FEATURE NORMALIZED =============================
total_active_days = (
    daily_device_connects
    .groupby('user')['only_date']
    .nunique()
    .reset_index(name='total_active_days')
)

# Merge with above_avg_daily_counter
above_avg_daily_ratio = above_avg_daily_counter.merge(total_active_days, on='user', how='left')
above_avg_daily_ratio['ratio_of_above_avg_days'] = (
    above_avg_daily_ratio['above_avg_daily_counter'] / above_avg_daily_ratio['total_active_days']
)

# Final cleanup: fill NaNs if any
above_avg_daily_ratio['ratio_of_above_avg_days'] = above_avg_daily_ratio['ratio_of_above_avg_days'].fillna(0)

# Merge with main features
features = features.merge(
    above_avg_daily_ratio[['user', 'ratio_of_above_avg_days']], on='user', how='left'
)
features['ratio_of_above_avg_days'] = features['ratio_of_above_avg_days'].fillna(0)

# ============================= RISK SCORE MODELLING ==========================
# 0 - 6
features['risk_score'] = (
	(features['late_hour_device_use'] > 3).astype(int) * 2 +
	(features['missing_disconnects'] > 5).astype(int) * 1 +
	(features['distinct_pcs_used'] > 1).astype(int) * 1 +
	(features['ratio_of_above_avg_days'] > 0.3).astype(int) * 2
)
# Add new rule: late_hour_device_use > 100 AND distinct_pcs_used > 100 → +1
features['risk_score'] += (
	((features['late_hour_device_use'] > 100) & (features['distinct_pcs_used'] > 100)).astype(int) * 1
)




# Final feature table
features.head()
```

```python
df_email = email_log.copy()
```

```python
# emails_sent_per_day
df_email['date_only'] = pd.to_datetime(df_email['date']).dt.date
emails_per_day = df_email.groupby(['user', 'date_only']).size().groupby('user').mean().reset_index(name='emails_sent_per_day')

# unique_recipients_count
df_email['to_list'] = df_email['to'].fillna('').str.split(';')
unique_recipients = df_email.explode('to_list').groupby('user')['to_list'].nunique().reset_index(name='unique_recipients_count')

# external_emails_ratio i.e. non @dtaa.com
df_email['to_list'] = df_email['to'].fillna('').str.split(';')
df_email_exploded = df_email.explode('to_list')
df_email_exploded['external'] = ~df_email_exploded['to_list'].str.contains('@dtaa.com', na=False)
external_ratio = df_email_exploded.groupby('user')['external'].mean().reset_index(name='external_emails_ratio')

#attachment_count_avg
attachment_avg = df_email.groupby('user')['attachments'].mean().reset_index(name='attachment_count_avg')

# emails_sent_off_hours 
df_email['hour'] = pd.to_datetime(df_email['date']).dt.hour
df_email['off_hours'] = (df_email['hour'] < 6) | (df_email['hour'] > 18)
off_hours_count = df_email.groupby('user')['off_hours'].sum().reset_index(name='emails_sent_off_hours')

# bcc_usage_rate
df_email['has_bcc'] = df_email['bcc'].notnull()
bcc_rate = df_email.groupby('user')['has_bcc'].mean().reset_index(name='bcc_usage_rate')



from functools import reduce

dfs = [emails_per_day, unique_recipients, external_ratio, attachment_avg,
       off_hours_count, bcc_rate]

final_features_email = reduce(lambda left, right: pd.merge(left, right, on='user', how='outer'), dfs)

#============== SCORING =========================
final_features_email['risk_score'] = (
	(final_features_email['emails_sent_per_day'] > 15).astype(int) * 1 +
	(final_features_email['unique_recipients_count'] > 100).astype(int) * 1 +
	(final_features_email['external_emails_ratio'] > 0.5).astype(int) * 1 +
	(final_features_email['attachment_count_avg'] > 0.5).astype(int) * 1 +
	(final_features_email['bcc_usage_rate'] > 0.5).astype(int) * 1 +
	(final_features_email['emails_sent_off_hours'] > 50).astype(int) * 1 
)


```

```python
df_file = file_log.copy()
```

```python
# Files Copied Per Day
# Step 1: Count files copied per user per day
df_file['date_only'] = pd.to_datetime(df_file['date']).dt.date
files_per_day = df_file.groupby(['user', 'date_only'])['filename'].count().reset_index(name='files_copied_per_day')

# Step 2: Calculate total files and total active days per user
total_files = files_per_day.groupby('user')['files_copied_per_day'].sum().reset_index(name='total_files_copied')
active_days = files_per_day.groupby('user')['date_only'].nunique().reset_index(name='active_days')

# Step 3: Join and compute average
file_avg = pd.merge(total_files, active_days, on='user')
file_avg['avg_files_copied_per_day'] = file_avg['total_files_copied'] / file_avg['active_days']

# Average File Copy Size
df_file['content_length'] = df_file['content'].str.len()
avg_file_size = df_file.groupby('user')['content_length'].mean().reset_index(name='avg_file_copy_size')

# Header Type Distribution
df_file['filetype'] = df_file['filename'].str.extract(r'\.([a-zA-Z0-9]+)$')

# Create one-hot encoded columns for each filetype
filetype_dummies = pd.get_dummies(df_file['filetype'], prefix='filetype')

# Merge with original DataFrame and group by user
df_with_dummies = pd.concat([df_file[['user']], filetype_dummies], axis=1)
header_distribution = df_with_dummies.groupby('user').sum().reset_index()

# After Hours File Copy Ratio
df_file['datetime'] = pd.to_datetime(df_file['date'])
df_file['hour'] = df_file['datetime'].dt.hour
df_file['is_after_hours'] = ~df_file['hour'].between(6, 18)

# Calculate ratio of after-hours accesses per user
after_hours_ratio = df_file.groupby('user')['is_after_hours'].mean().reset_index(name='after_hours_file_copy_ratio')

dfs = [file_avg, header_distribution, after_hours_ratio]

final_features_file = reduce(lambda left, right: pd.merge(left, right, on='user', how='outer'), dfs)

final_features_file['risk_score'] = (
	(final_features_file['avg_files_copied_per_day'] > 15).astype(int) * 1 +
	(final_features_file['filetype_exe'] > 1).astype(int) * 2 +
	(final_features_file['filetype_zip'] > 1).astype(int) * 1 +
	(final_features_file['after_hours_file_copy_ratio'] > 0.1).astype(int) * 1 +
	(final_features_file['after_hours_file_copy_ratio'] > 0.8).astype(int) * 1 
)

```

```python
# Average Total URLs Visited Per Day
df_http['date_only'] = pd.to_datetime(df_http['date']).dt.date
daily_urls = df_http.groupby(['user', 'date_only']).size().reset_index(name='urls_per_day')

# Then average per user
avg_urls_per_day = daily_urls.groupby('user')['urls_per_day'].mean().reset_index(name='avg_urls_per_day')

# Malicious urls info
malicious_df = df_http[df_http['malicious_url_flag'] == 1]

# 1. Total malicious URLs per user
total_malicious_urls = malicious_df.groupby('user').size().reset_index(name='total_malicious_urls')

# 2. Malicious URLs per day per user
malicious_per_day = malicious_df.groupby(['user', 'date_only']).size().reset_index(name='malicious_urls_per_day')

# 3. Average malicious URLs per day per user
avg_malicious_per_day = malicious_per_day.groupby('user')['malicious_urls_per_day'].mean().reset_index(name='avg_malicious_urls_per_day')

# 4. Merge both into a single dataframe
#malicious_features = pd.merge(total_malicious_urls, avg_malicious_per_day, on='user', how='outer').fillna(0)

# Distinct Domains Visited
from urllib.parse import urlparse

df_http['domain'] = df_http['url'].apply(lambda x: urlparse(x).netloc)
daily_distinct_domains = df_http.groupby(['user', 'date_only'])['domain'].nunique().reset_index(name='distinct_domains_per_day')

# Average distinct domains per day per user
avg_distinct_domains = daily_distinct_domains.groupby('user')['distinct_domains_per_day'].mean().reset_index(name='avg_distinct_domains_per_day')

# After-Hours Browsing Count
df_http['hour'] = pd.to_datetime(df_http['date']).dt.hour
df_http['after_hours'] = (df_http['hour'] < 6) | (df_http['hour'] > 18)
after_hours_daily = df_http[df_http['after_hours']].groupby(['user', 'date_only']).size().reset_index(name='after_hours_count')

# Then average per user
avg_after_hours = after_hours_daily.groupby('user')['after_hours_count'].mean().reset_index(name='avg_after_hours_browsing_per_day')

# List of all individual feature DataFrames
feature_dfs = [avg_urls_per_day, avg_malicious_per_day, total_malicious_urls, avg_distinct_domains, avg_after_hours]

# Merge all on 'user'
final_features_http = reduce(lambda left, right: pd.merge(left, right, on='user', how='outer'), feature_dfs)

final_features_http['risk_score'] = (
	(final_features_http['avg_after_hours_browsing_per_day'] > 1).astype(int) * 1 +
    	(final_features_http['avg_malicious_urls_per_day'] > 3).astype(int) * 2 +
    	(final_features_http['total_malicious_urls'] > 50).astype(int) * 2 +
	(final_features_http['total_malicious_urls'] > 100).astype(int) * 1
)




# Optionally fill NaNs if needed
final_features_http.fillna(0, inplace=True)
```

LOGON (log in and out info i.e. session)

```python
logon_log.head(15)
```

```python
df_logon = logon_log.copy()
```

```python
# Convert date to datetime and extract relevant parts
df_logon['datetime'] = pd.to_datetime(df_logon['date'])
df_logon['date_only'] = df_logon['datetime'].dt.date
df_logon['time'] = df_logon['datetime'].dt.time
df_logon['hour'] = df_logon['datetime'].dt.hour
df_logon['weekday'] = df_logon['datetime'].dt.weekday  # 0 = Monday, 6 = Sunday

# Filter Logon and Logoff separately
df_logon_only = df_logon[df_logon['activity'] == 'Logon']
df_logoff_only = df_logon[df_logon['activity'] == 'Logoff']

# Average logon time per user (in hours)
avg_logon_time = df_logon_only.groupby('user')['datetime'].apply(
    lambda x: pd.to_timedelta(x.dt.time.astype(str)).mean().total_seconds() / 3600
).reset_index(name='avg_logon_time_hours')

# Average logoff time per user (in hours)
avg_logoff_time = df_logoff_only.groupby('user')['datetime'].apply(
    lambda x: pd.to_timedelta(x.dt.time.astype(str)).mean().total_seconds() / 3600
).reset_index(name='avg_logoff_time_hours')

# Logon time standard deviation (in hours)
logon_std_dev = df_logon_only.groupby('user')['datetime'].apply(
    lambda x: pd.to_timedelta(x.dt.time.astype(str)).std().total_seconds() / 3600
).reset_index(name='logon_std_dev_hours')

## after_hours_logons & outside_working_hours_logons
after_hours = (df_logon['activity'] == 'Logon') & ((df_logon['hour'] < 6) | (df_logon['hour'] > 18))
after_hours_logons = df_logon[after_hours].groupby('user').size().reset_index(name='after_hours_logons')

### weekend_logons
weekend_logons = df_logon[
    (df_logon['activity'] == 'Logon') & (df_logon['weekday'] >= 5)
].groupby('user').size().reset_index(name='weekend_logons')

#
# Most frequently used PC per user
most_used_pc = df_logon_only.groupby('user')['pc'].agg(lambda x: x.mode().iloc[0]).reset_index()
most_used_pc.columns = ['user', 'main_pc']

# Merge back and compare
df_logon_only = df_logon_only.merge(most_used_pc, on='user', how='left')
df_logon_only['other_pc'] = df_logon_only['pc'] != df_logon_only['main_pc']

logons_on_other_user_pcs = df_logon_only[df_logon_only['other_pc']].groupby('user').size().reset_index(name='logons_on_other_user_pcs')



feature_dfs = [
    avg_logon_time, avg_logoff_time, logon_std_dev,
    after_hours_logons, weekend_logons,
    logons_on_other_user_pcs
]

df_logon_features = reduce(lambda left, right: pd.merge(left, right, on='user', how='outer'), feature_dfs)

df_logon_features['risk_score'] = (
	(df_logon_features['logon_std_dev_hours'] > 2).astype(int) * 1 +
	(df_logon_features['after_hours_logons'] > 1).astype(int) * 2 +
	(df_logon_features['weekend_logons'] > 1).astype(int) * 1 +
	(df_logon_features['logons_on_other_user_pcs'] > 1).astype(int) * 2 
)



df_logon_features.fillna(0, inplace=True)
```

```python
# Apply distributed weights before joining the processed log files
device_features['risk_score'] = device_features['risk_score']*0.1
final_features_email['risk_score'] = final_features_email['risk_score']*0.2
final_features_file['risk_score'] = final_features_file['risk_score']*0.2
final_features_http['risk_score'] = final_features_http['risk_score']*0.2
df_logon_features['risk_score'] = df_logon_features['risk_score']*0.3
```

```python
# List of DataFrames to merge
feature_dfs = [
    device_features,
    final_features_email,
    final_features_file,
    final_features_http,
    df_logon_features
]

# Step 1: Rename 'risk_score' in each DataFrame before merging to avoid conflicts
for i, df in enumerate(feature_dfs):
    df.rename(columns={'risk_score': f'risk_score_{i+1}'}, inplace=True)

# Step 2: Merge all DataFrames on 'user'
all_features = reduce(lambda left, right: pd.merge(left, right, on='user', how='outer'), feature_dfs)

# Step 3: Fill NaNs with 0
all_features.fillna(0, inplace=True)

# Step 4: Sum all renamed risk scores into a new column
risk_score_cols = [col for col in all_features.columns if col.startswith('risk_score_')]
all_features['total_risk_score'] = all_features[risk_score_cols].sum(axis=1)
```

```python
all_features['potential_threat'] = (
    (all_features['total_risk_score'] >= 3.4)
)
```
